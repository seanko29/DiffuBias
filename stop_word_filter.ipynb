```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


csv_file = '/home/seanko/aaai/save/blip_texts/blip_bar_aaai/5pct/bar_5pct_image_captioning_t5_XXL.csv'
df = pd.read_csv(csv_file)

# nltk.download('punkt')
# nltk.download('stopwords)
# Get the English stop words from NLTK
stop_words = set(stopwords.words('english'))

# Initialize a dictionary to count occurrences of non-stop words
word_counts = {}

# Iterate over each row in the DataFrame
for index, feature in df.iterrows():
    # print(feature[0])
    label = feature[0].split('/')[-2]

    # Iterate over each text feature in the row (assuming there are 3 text columns)
    for text in [feature[1], feature[2], feature[3]]:
        # Tokenize the text
        word_tokens = word_tokenize(text)

        # Filter out the tokens that are not stop words and count them
        for word in word_tokens:
            word_lower = word.lower()
            if word_lower not in stop_words and word_lower.isalpha():
                word_counts[word_lower] = word_counts.get(word_lower, 0) + 1
                
# Convert the dictionary to a DataFrame
words_df = pd.DataFrame(list(word_counts.items()), columns=['Word', 'Count'])

# Sort the DataFrame by count in descending order
words_df.sort_values(by='Count', ascending=False, inplace=True)

# Optionally, save the DataFrame to a CSV file
words_df.to_csv('word_counts.csv', index=False)
                
print(word_counts)

# Sort the dictionary by count in descending order
sorted_word_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)

# Print the sorted dictionary
for word, count in sorted_word_counts:
    print(f"{word}: {count}")
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Ensure the 'punkt' tokenizer is downloaded
# nltk.download('punkt')

# Define the number of classes
number_of_classes = 6
top_n_threshold = 2 * number_of_classes

# Read your original dataframe
csv_file = '/home/seanko/aaai/save/blip_texts/blip_bar_aaai/5pct/bar_5pct_image_captioning_t5_XXL.csv'
df = pd.read_csv(csv_file)

# Get the English stop words from NLTK
stop_words = set(stopwords.words('english'))

# Initialize a dictionary to count occurrences of non-stop words
word_counts = {}

# Tokenize and count words in the dataframe
for index, row in df.iterrows():
    label = row[0].split('/')[-2]
    for text in [row[1], row[2], row[3]]:  # Replace with your actual text columns
        word_tokens = word_tokenize(text)
        for word in word_tokens:
            word_lower = word.lower()
            if word_lower not in stop_words and word_lower.isalpha():
                word_counts[word_lower] = word_counts.get(word_lower, 0) + 1

# Sort and get the top N words
top_n_words = [word for word, count in sorted(word_counts.items(), key=lambda item: item[1], reverse=True)[:top_n_threshold]]

# Filter rows in the original dataframe and concatenate texts
filtered_data = []
for index, row in df.iterrows():
    label = row[0].split('/')[-2]
    # print(label)
    row_texts = [row[1], row[2], row[3]]  # Replace with your actual text columns
    # concatenated_texts = " ".join(row_texts).lower()
    for sentence in row_texts:
    # Check if any of the top N words are in the concatenated texts
        if any(word in sentence for word in top_n_words):
            filtered_data.append([label, sentence])

# Create a new dataframe with the filtered data
filtered_df = pd.DataFrame(filtered_data, columns=['Label', 'GeneratedTexts'])

# Save the filtered dataframe to a new CSV file
filtered_csv_file = 'filtered_csv_file.csv'
filtered_df.to_csv(filtered_csv_file, index=False)
```
